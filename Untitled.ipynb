```python
import numpy as np
import math
import cv2
import imutils
import serial
from time import sleep
dup = []
k="str"
value = serial.Serial('COM26', 9600, timeout=2)
'''lower = {'green': (60, 96, 25)}
upper = {'green': (80, 255, 255)}'''
'''lower = {'green': (60, 39, 107), 'yellow': (27, 45, 137), 'orange': (0, 89, 142),'pink':(155,119,110)}
upper = {'green': (80, 255, 255), 'yellow': (55, 255, 255), 'orange': (10, 255, 255),'pink':(179,255,255)}'''
lower = {'orange': (0, 89, 255)}
upper = { 'orange': (25, 255, 255)}
camera = cv2.VideoCapture(1)
def convert(x,y):
    k='k'
    x=str(x)
    y=str(y)
    x+=k
    value.write(x.encode())
    sleep(1)
    value.write(y.encode())
    x=value.readline().decode('ascii')
    y=value.readline().decode('ascii')
    print("The values are sent to the bot")
    print(x)
def calculate():
    sleep(0.5)
    distance=value.readline().decode('ascii')
    distance=int(distance)
    print("The distance of the slave bot from the master bot is:",distance)
    sleep(0.5)
    angle=value.readline().decode('ascii')
    angle=int(angle)
    print("the angle w.r.t. y axis is",angle)
    if angle>=90:
        angle=angle-90
    else:
        angle=90-angle
        angle=-angle
    print("the angle w.r.t.x axis is",angle)
    x1=distance*math.cos((angle*(3.142/180)))
    y1=distance*math.sin((angle*(3.142/180)))
    x1=round(x1);
    y1=round(y1);
    print(x1,y1)
    convert(x1,y1)
def serialcolor(h):
    value.write(h.encode())
    sleep(3)
    g=value.readline().decode('ascii')
    print("value is sent",g)
    calculate()
def inputvalues(x):
    
    value.write(x.encode())
    #sleep(2)
   # value.write(y.encode())
#g='k'

x=(input("x coordinate"))
#y=(input("y coordinate"))
#x+=g
inputvalues(x)
while True:
    _, frame = camera.read()
    roi = frame[0:480,330:340]
    hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
    for key, _ in upper.items():
        lower_blue = np.array(lower[key])
        upper_blue = np.array(upper[key])
        mask = cv2.inRange(hsv, lower_blue, upper_blue)
        blurred = cv2.GaussianBlur(frame, (5, 5), 0)
        erode_element = cv2.getStructuringElement(cv2.MORPH_RECT, (6, 6))
        dilate_element = cv2.getStructuringElement(cv2.MORPH_RECT, (6, 6))
        eroded_mask = cv2.erode(mask, erode_element)
        dilated_mask = cv2.dilate(eroded_mask, dilate_element)
        cnts = cv2.findContours(dilated_mask.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
        cnts = imutils.grab_contours(cnts)
        center = None
        if len(cnts) > 0:
            # find the largest contour in the mask, then use
            # it to compute the minimum enclosing circle and
            # centroid
            c = max(cnts, key=cv2.contourArea)
            ((x, y), radius) = cv2.minEnclosingCircle(c)
            M = cv2.moments(c)


            # only proceed if the radius meets a minimum size
            if radius > 20:
                # draw the circle and centroid on the frame,
                # then update the list of tracked points
                cv2.circle(roi, (int(x), int(y)), int(radius),
                           (0, 255, 255), 2)
                cv2.circle(roi, center, 5, (0, 0, 255), -1)

                if key not in dup:
                    print(key)
                    serialcolor(key)
                    dup.append(key)  
                
        cv2.imshow("output1", mask)
        cv2.imshow("output2", roi)

        
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
 
camera.release()

cv2.destroyAllWindows()
 
```
